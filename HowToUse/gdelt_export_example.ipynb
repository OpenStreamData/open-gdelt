#逐行输出，检查数据格式

```python
import requests
import json

def print_kv(data, indent=0):
    prefix = "  " * indent
    if isinstance(data, dict):
        for k, v in data.items():
            if isinstance(v, (dict, list)):
                print(f"{prefix}{k}:")
                print_kv(v, indent + 1)
            else:
                print(f"{prefix}{k}: {v}")
    elif isinstance(data, list):
        for i, item in enumerate(data):
            print(f"{prefix}[{i}]")
            print_kv(item, indent + 1)
    else:
        print(f"{prefix}{data}")

url = "https://open-gdelt.openstreamdata.top/export.jsonl.txt"

with requests.get(url, stream=True, timeout=30) as r:
    r.raise_for_status()
    for idx, line in enumerate(r.iter_lines(decode_unicode=True), 1):
        if not line:
            continue

        obj = json.loads(line)
        print(f"\n--- 第 {idx} 行 ---")
        print_kv(obj)
```

# 快速导出为Excel表


```python
import requests
import json
import csv

url = "https://open-gdelt.openstreamdata.top/export.jsonl.txt"
csv_file = "output.csv"

rows = []

with requests.get(url, stream=True, timeout=30) as r:
    r.raise_for_status()
    for line in r.iter_lines(decode_unicode=True):
        if line:
            rows.append(json.loads(line))

fieldnames = rows[0].keys()

with open(csv_file, "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(rows)

print(f"已保存到 {csv_file}")
```
